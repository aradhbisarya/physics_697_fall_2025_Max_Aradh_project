```python
import numpy as np

m = 1
w = 1
N = 24
F = 1
C = 3

class term:
    def __init__(self):
        self.coeff = 1
        self.ops = []
    def __mul__(self, other):
        result = term()
        result.coeff = self.coeff * other.coeff
        result.ops = self.ops + other.ops
        return result

class sop_state:
    def __init__(self,):
        self.coeff = 0
        self.terms = []
    def __mul__(self, other):
        result = sop_state()
        result.coeff = self.coeff * other.coeff
        for term1 in self.terms:
            for term2 in other.terms:
                result.terms.append(term1*term2)
        return result

def l(n,i):
    return ((n-1)*F*C) + i - 1

def Q_ij(i,j,n):
    result = term()
    result.coeff = 1
    result.ops.append(('Q', l(n,i), l(n,j)))
    return result

def Q(i,n,m):
    result = sop_state()
    result.coeff = 1
    result.terms.append(Q_ij(i,j,n))
    return result
H_electric = sop_state()
H_electric.coeff = 1
```
